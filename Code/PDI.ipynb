```python
import numpy as np
import skimage.io
import skimage.util
from matplotlib import pyplot as plt

ft_class = ["Apple", "Banana", "Pineapple", "Orange", "Pear", "Kiwi_fruit", "Avocado", "Dragon_fruit", "Lemon", "Melon"]
ft_version = ['V1', 'V2']
ft_color = ['B', 'W']

fig, axes = plt.subplots(5, 2, figsize=(12, 12))  # Alterar para 5 linhas e 2 colunas para acomodar 12 classes
ax = axes.ravel()

for x in range(len(ft_class)):
    img_fruit = []
    for ver in ft_version:
        for col in ft_color:
            img_fruit.append(skimage.io.imread(f'FrutasORG/{ft_class[x]}/0-{str(x+1).zfill(2)}-{ver}-{col}.png'))  # Ajustar o nmero para comear em 0
    m = np.concatenate(img_fruit, axis=1)
    ax[x].imshow(m, cmap='gray')
    ax[x].set_axis_off()
    ax[x].set_title(f"{ft_class[x]}")

fig.tight_layout()
plt.show()
```
